```python
import pandas as pd

# Load the file into a pandas DataFrame
file_path = 'results/homer_gene_annotations.txt'  # Adjust this path
df = pd.read_csv(file_path, sep='\t', header=0)  # Assuming the file has a header

# Extract the 'Gene Name' column
gene_names = df['Gene Name'].dropna().unique()

# Print or save the gene names
for name in gene_names:
    print(name)
```

```python
output_file = 'results/gene_names.txt'
with open(output_file, 'w') as f:
    for name in gene_names:
        f.write(name + '\n')
```
